```r
library(httr)
library(ggplot2)

# 웹에서 파이의 데이터를 읽어오기
url <- "https://www.angio.net/pi/digits/pi1000000.txt"
response <- GET(url)
pi_data <- content(response, "text", encoding = "UTF-8")

# '3.'을 제외하고 숫자 부분만 추출
pi_digits <- substr(pi_data, 3, 1000002)

# 숫자별로 분리하여 벡터 생성
digits_vector_1 <- unlist(strsplit(pi_digits, ""))

# 각 단일 숫자의 빈도수 계산 및 상대 빈도 계산
digit_frequencies <- table(digits_vector_1)  #1) 각 숫자의 빈도수를 구하라.

# 빈도 데이터 프레임 생성
df <- data.frame(Digit = names(digit_frequencies), Frequency = as.numeric(digit_frequencies))

# 기대빈도 계산 (총 숫자 개수를 10으로 나눈 값)
total_digits <- length(digits_vector_1)
expected_frequency <- total_digits / 10

# 빈도/기대빈도 데이터 프레임 생성
df$ExpectedFrequency <- expected_frequency
df$FrequencyToExpected <- df$Frequency / expected_frequency

# 그래프 볼 화면
dev.new()
############## 히스토그램 그리기_빈도
hist(df$Frequency, breaks=1000,main='한 자리 숫자 (0~9)',xlab='빈도',ylab='Frequency')
############# 히스토그램 그리기_상대빈도
hist(df$FrequencyToExpected, breaks=1000,main='한 자리 숫자 (0~9)',xlab='빈도/기대빈도',ylab='Frequency')


#####################(2) 2개씩 잘라서 빈도수####################

# 숫자를 두 개씩 분리하여 벡터 생성
digits_vector_2 <- sapply(1:(nchar(pi_digits)/2), function(i) substr(pi_digits, (2*i)-1, 2*i))

# 각 두 자리 숫자의 빈도수 계산
digit_frequencies <- table(digits_vector_2)

# 빈도 데이터 프레임 생성
df <- data.frame(Digit = names(digit_frequencies), Frequency = as.numeric(digit_frequencies))

# Frequency 열의 최대값과 최소값에 해당하는 Digit 값
max_digit <- df$Digit[which.max(df$Frequency)]
min_digit <- df$Digit[which.min(df$Frequency)]

# 결과 출력
cat("Frequency의 최대값:", max(df$Frequency), ", Digit 값:", max_digit, "\n")
cat("Frequency의 최소값:", min(df$Frequency), ", Digit 값:", min_digit, "\n")

# 기대빈도 계산 (총 숫자 쌍 개수를 100으로 나눈 값)
total_pairs <- length(digits_vector_2)
expected_frequency <- total_pairs / 100

# 빈도/기대빈도 데이터 프레임 생성
df$ExpectedFrequency <- expected_frequency
df$FrequencyToExpected <- df$Frequency / expected_frequency

# 그래프 볼 화면
dev.new()
############## 히스토그램 그리기_빈도
hist(df$Frequency, breaks=1000, main='두 자리 숫자 (00~99)', xlab='빈도', ylab='Frequency')
############# 히스토그램 그리기_빈도/기대빈도
hist(df$FrequencyToExpected, breaks=1000, main='두 자리 숫자 (00~99)', xlab='빈도/기대빈도', ylab='Frequency')

#####################(2) 3개씩 잘라서 빈도수####################

# 숫자를 세 개씩 분리하여 벡터 생성
digits_vector_3 <- sapply(1:(nchar(pi_digits)/3), function(i) substr(pi_digits, (3*i)-2, 3*i))

# 각 세 자리 숫자의 빈도수 계산
digit_frequencies <- table(digits_vector_3)

# 빈도 데이터 프레임 생성
df <- data.frame(Digit = names(digit_frequencies), Frequency = as.numeric(digit_frequencies))

# Frequency 열의 최대값과 최소값
max_freq_value <- max(df$Frequency)
min_freq_value <- min(df$Frequency)

# 최대값과 최소값에 해당하는 모든 Digit 값
max_digits <- df$Digit[df$Frequency == max_freq_value]
min_digits <- df$Digit[df$Frequency == min_freq_value]

# 결과 출력
cat("Frequency의 최대값:", max_freq_value, ", Digit 값:", paste(max_digits, collapse=", "), "\n")
cat("Frequency의 최소값:", min_freq_value, ", Digit 값:", paste(min_digits, collapse=", "), "\n")


# 기대빈도 계산 (총 숫자 삼중조 개수를 1000으로 나눈 값)
total_triples <- length(digits_vector_3)
expected_frequency <- total_triples / 1000

# 빈도/기대빈도 데이터 프레임 생성
df$ExpectedFrequency <- expected_frequency
df$FrequencyToExpected <- df$Frequency / expected_frequency

# 그래프 볼 화면
dev.new()
############## 히스토그램 그리기_빈도
hist(df$Frequency, breaks=1000, main='세 자리 숫자 (000~999)', xlab='빈도', ylab='Frequency')
############# 히스토그램 그리기_빈도/기대빈도
hist(df$FrequencyToExpected, breaks=1000, main='세 자리 숫자 (000~999)', xlab='빈도/기대빈도', ylab='Frequency')


#######################################################
#4) 100만 개의 숫자 중에서 0~4까지 연속으로 가장 길게 나온 개수를 구하라.

# 0부터 9까지 각 숫자의 최대 연속 길이와 그 길이가 나타난 횟수를 저장할 벡터 초기화
max_lengths <- numeric(10)
length_counts <- numeric(10)

# 현재 연속 길이를 저장할 임시 벡터 초기화
current_lengths <- numeric(10)

# pi_digits의 각 자리 숫자를 순회하며 확인
for (i in 1:nchar(pi_digits)) {
  digit <- as.numeric(substr(pi_digits, i, i))

  # 모든 숫자의 연속 길이를 1 증가, 현재 숫자만 제외하고 0으로 초기화
  current_lengths <- ifelse(0:9 == digit, current_lengths + 1, 0)

  # 각 숫자에 대해 최대 연속 길이와 그 길이가 나타난 횟수 업데이트
  for (j in 1:10) {
    if (current_lengths[j] > max_lengths[j]) {
      max_lengths[j] <- current_lengths[j]
      length_counts[j] <- 1
    } else if (current_lengths[j] == max_lengths[j] && max_lengths[j] != 0) {
      length_counts[j] <- length_counts[j] + 1
    }
  }
}

# 결과 출력
for (digit in 0:9) {
  cat(sprintf("%d이 연속으로 가장 길게 나온 개수 : %d)\n", digit,max_lengths[digit + 1]))
}

##############################################################
#######5) 3이 연속으로 가장 길게 나오는 위치

current_length <- 0
max_length <- 0

# 최대 연속 길이의 시작 위치
start_position <- 0
max_start_position <- 0

for (i in 1:length(digits_vector_1)) {
  if (digits_vector_1[i] == 3) {
    current_length <- current_length + 1
    if (current_length == 1) { # 새로운 3의 연속 시작
      start_position <- i
    }
    if (current_length > max_length) {
      max_length <- current_length
      max_start_position <- start_position
    }
  } else {
    current_length <- 0
  }
}

# 결과 출력
if (max_length > 0) {
  cat(sprintf("3이 연속으로 가장 길게 나타나는 첫 위치: %d, 길이: %d\n", max_start_position, max_length))
} else {
  cat("3이 연속으로 나타나지 않습니다.\n")
}
```
